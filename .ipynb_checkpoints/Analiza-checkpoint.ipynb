```python
%matplotlib inline
import pandas as pd

from IPython.core.display import HTML
css = open('style-table.css').read() + open('style-notebook.css').read()
HTML('<style>{}</style>'.format(css))
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16, 8]
```

```python
# Dane są podzielone na lata, w celu ich zbiorczej analizy trzeba było połączyc wszystkie pliki w jedną ramke danych
```

```python
import os

consumption = pd.DataFrame()
for filename in os.listdir('./data'):
    print(filename)
    tempDf = pd.read_csv('data/'+filename)
    consumption = consumption.append(tempDf)
```

```python
consumption.head()
```

```python
c = consumption
c.YEAR = pd.to_numeric(c.YEAR, errors='coerce')
c.dropna(inplace=True)
c
```

```python
g = c.groupby(['YEAR','MAKE']).mean()
# Pivot tabeli -> przesunięcie producenta jako kolumny
g = g.unstack('MAKE')
# Sortowanie po roku z którego są dane
g = g.sort_index()
# Pominięcie producentów z brakującymi danymi w przedziale 2000-2017
g = g.dropna(axis=1)

g.plot(y='HWY', title='Śr. użycie paliwa w mieście w wszystkich modelach producenta z danego roku',
       legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

```

```python
# widać bardzo słaby trend ogólny zmiejszania się zużycia palowa.
# ciekawa zależność jest widoczna dopiero kiedy podzielimy ten wykres na 2 czesci
```

```python
g = c[c.YEAR > 2008].groupby(['YEAR','MAKE']).mean()
# Pivot tabeli -> przesunięcie producenta jako kolumny
g = g.unstack('MAKE')
# Sortowanie po roku z którego są dane
g = g.sort_index()
# Pominięcie producentów z brakującymi danymi w przedziale 2008-2017
g = g.dropna(axis=1)

g.plot(y='CITY', title='Śr. użycie paliwa w mieście w wszystkich modelach producenta lata 2009-2017',
       legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
g = c[c.YEAR <= 2008].groupby(['YEAR','MAKE']).mean()
# Pivot tabeli -> przesunięcie producenta jako kolumny
g = g.unstack('MAKE')
# Sortowanie po roku z którego są dane
g = g.sort_index()
# Pominięcie producentów z brakującymi danymi w przedziale 2000-2008
g = g.dropna(axis=1)

g.plot(y='CITY', title='Śr. użycie paliwa w mieście w wszystkich modelach producenta lata 2000-2008',
       legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
# widać że w latach 2000-2007 nie istniał żaden widoczny trend wśród producentów samochodów,
# do obniżania zużycia paliwa w mieście.
# dopiero przez ostatnie 10-12 lat widać wyraźny trend spadkowy
```

```python
# Jest to mocno widoczne kiedy zarysujemy jedną linie tredu ze wszystkich producentów
```

```python
g = c[c.YEAR > 2008].groupby(['YEAR']).mean()
# Sortowanie po roku z którego są dane
g = g.sort_index()
# Pominięcie producentów z brakującymi danymi w przedziale 2009-2017
g = g.dropna(axis=1)

g.plot(y='CITY', title='Śr. użycie paliwa w mieście w latach 2009-2017',
       legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))


# Wykres ma widoczna linie trendu
```

```python
g = c[c.YEAR <= 2008].groupby(['YEAR']).mean()
# Sortowanie po roku z którego są dane
g = g.sort_index()
# Pominięcie producentów z brakującymi danymi w przedziale 2000-2008
g = g.dropna(axis=1)

g.plot(y='CITY', title='Śr. użycie paliwa w mieście w latach 2000-2008',
       legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))


# Brak jednoznaczengo trendu
```

```python
# Druga cześć analizy
Które modele mają największą różnice pomiedzy ich średnim spalaniem w mieście, a w trasie.
```

```python
c = consumption
diff = c.CITY - c.HWY
c['diff'] = diff
c = c.sort_values('diff')
c.head(20)

# W tabeli widzimy że w przypadku samochodów hybrydowych mamy ciekawą zależność
# W niektórych modelach, ich zużycie paliwa w mieście jest mniejsze niż w trasie
# Nie oznacza to jednak że ich żużycie w trasie jest duże.
# Ich technologia hybrydowa jest jednak przydatna jedynie podczas jazdy w mieście
```
